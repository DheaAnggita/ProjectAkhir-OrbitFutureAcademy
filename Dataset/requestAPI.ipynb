```python
import json
import requests
import datetime

from google.colab import drive
drive.mount("/content/drive")

import os
os.chdir('/content/drive/MyDrive/Orbit')

tahun = 2011
while tahun <= 2021:
    start_date = datetime.date(tahun, 1, 1)
    end_date = datetime.date(tahun, 12, 31)
    delta = datetime.timedelta(days=1)
    stahun = str(tahun)

    while start_date <= end_date:

        # print(start_date)
        start = start_date.strftime("%Y-%m-%d")
        request = requests.get('https://api.worldweatheronline.com/premium/v1/past-weather.ashx?key=c5b866d276d74c578a363815221305&q=-7.784680,110.430673&date='+start+'&enddate='+start+'&format=json&tp=24')
        request_text = request.text

        data = json.loads(request_text)

        data_serialized = json.dump(data, open('/content/drive/MyDrive/Orbit/PA/Dataset/Maguwoharjo'+stahun+'.json', "a"), indent = 4)
        
        start_date += delta
    
    tahun += 1
```
